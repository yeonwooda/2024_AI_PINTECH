```python
# 선형 회귀
import numpy as np

perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0,
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5,
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5,
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0,
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0,
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight)
```

```python
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
print(train_input.shape, test_input.shape)
```

```python
# 훈련
from sklearn.neighbors import KNeighborsRegressor #회귀 모델
knr = KNeighborsRegressor(n_neighbors=3) # 최근접 이웃의 개수 기본값 5가 아닌 3으로 변경 - 모델의 복잡도가 높아진다

knr.fit(train_input, train_target)
```

```python
knr.predict([[50], [100]]) # 예측에서 벗어나도 가장 가까운 친구들 끼리 평균내서 값을 냄
```

```python
distances, indexes = knr.kneighbors([[50]]) # 50근처에 있는 이웃 구하기

# 넘파이 평균 mean
np.mean(train_target[indexes]) # 50을 둘러싸고 있는 이웃의 평균
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker='D') # 50cm 길이의 농어 최근접 이웃
plt.scatter(50, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 선형 회귀
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

# 선형 회귀 모델 훈련
lr.fit(train_input, train_target)

# 50cm 농어에 대한 예측
print(lr.predict([[50]]))
```

```python
# 최적의 기울기 - 계수, 가중치 - lr.coef_ (가중치, 기울기)
# 최적의 절편 - lr.intercept_

print(lr.coef_, lr.intercept_) # y = 36.8x - 630.96 : x - 길이, y - 무게
```

```python
# matplotlib - plot : 직선, 곡선 그래프

plt.scatter(train_input, train_target) # 훈련세트

#1차 방정식 그래프
plt.plot([8, 50], [8*lr.coef_[0]+lr.intercept_, 50*lr.coef_[0]+lr.intercept_])
plt.scatter(50, 1220, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(lr.score(train_input, train_target))  # 훈련 세트 결정 계수
print(lr.score(test_input, test_target))    # 테스트 세트 결정 계수
```
