```python
import requests

response = requests.get("https://news.naver.com/")
html = response.text
```

```python
print(html)
```

```python
import re
"""
<img class="tv_program_img" loading="lazy" src="https://mimgnews.pstatic.net/image/upload/office_logo/056/2022/12/29/logo_056_67_20221229163228.jpg" width="117" alt="" data-pressname="KBS" onerror="showNoImage(this, 'pressname')">
"""
p = re.compile(r"<img[^>]+>")
imgs = p.findall(html)
for img in imgs:
    print(img)
```

```python
import re
'''
<span class="Nitem_link_menu">언론사별</span>
'''
p = re.compile(r"<span.*class=['\"]Nitem_link_menu['\"][^>]*>[^<]+</span>")
items = p.findall(html)
for item in items:
    print(item)
```

```python
import re
'''
<span class="Nitem_link_menu">언론사별</span>
'''
p = re.compile(r"<span.*class=['\"]Nitem_link_menu['\"][^>]*>[^<]+</span>")
itr = p.finditer(html)
```

```python
for m in itr:
    print("-" * 50)
    print("group():", m.group())
    print("start():", m.start())
    print("end():", m.end())
    print("span():", m.span())
    #print("-" * 50)
```
